```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
agsel = pd.read_csv("car_resale_prices.csv")
agsel.head()
```

### Kita import semua library yang kita mau pakai, setelah itu kita panggil csv nya dan kita tampilkan


```python
agsel.set_index('full_name', inplace=True)
agsel.head()
```

```python
resale = agsel["body_type"].value_counts()
```

resale sebagai variabel yang menyimpan data frame berdasarkan kolom body_type lalu menghitung berapa banyak target yang didapatkan

```python
resale
```

Menampilkan banyaknya data dari suatu target

```python
agsel['registered_year'] = pd.to_numeric(agsel['registered_year'], errors='coerce')
```

karena disini data saya kebanyakan object jadi saya menconvert menjadi integer

```python
mean_data = agsel.groupby("body_type")[["registered_year"]].mean()
mean_data
```

```python
filtered_data = agsel[agsel['registered_year'] >= 1500]

mean_data = filtered_data.groupby("body_type")[["registered_year"]].mean()

x = mean_data.index
y = mean_data["registered_year"]
plt.figure(figsize=(12, 6))
plt.bar(x, y)
plt.xticks(rotation=90)
plt.title('Rata-rata Tahun Rilis')
plt.xlabel("Tipe BodI")
plt.ylabel("Rata-rata Tahun Rilis")
plt.ylim(1500, max(y) + 100)
plt.show()
```

***DIAGRAM BATANG***

Menampilkan seberapa banyak dan membandingkan tahun rilis yang didapatkan dari data set yang ada, dari data tersebut terlihat bahwa rata rata mobil di rilis pada tahun 2000an

```python
plt.figure(figsize=(8,8))
explode_pie = [0.1, 0.1, 0.1, 0.1,0.3, 0.7, 0.8, 0.9, 1, 1.1, 1.3, 1.5, 1.7, 1.9, 2.1, 2.3, 2.5, 2.7, 2.9, 3.1, 3.3, 3.5, 3.7, 3.9, 4.1,4.3 ]
plt.pie(resale, labels=resale.index,explode=explode_pie, startangle=90,autopct = "%.5f%%")
plt.show()
```

***LINE PLOT***

mengunakan line plot untuk melihat membandingkan rata-rata setiap bahan bakar dibeli pada tahun kapan

```python
kolom = ['registered_year', 'seats', 'engine_capacity']

korelasi = agsel[kolom].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(korelasi, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap Korelasi Kolom Numerik')
plt.show()
```

***HEATMAP***

memvisualisasikan korelasi antara kolom-kolom yang bersifat numerik
