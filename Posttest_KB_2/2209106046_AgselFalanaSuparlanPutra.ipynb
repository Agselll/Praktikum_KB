ketiga
tampilkan 10 record pertama dari dataset dengan attribute bertipe angka

```python
data = {}

data['Jumlah Record'] = Agsel.shape[0]
data['Jumlah Attribute'] = Agsel.shape[1]
data['Nama Attribute'] = Agsel.columns.tolist()
data['Jumlah Attribute Dengan Tipe Angka'] = Agsel.select_dtypes(include=['int64', 'float64']).shape[1]

dictKeys = list(data.keys())

table = prtl.PrettyTable()
table.add_column(dictKeys[0], [data['Jumlah Record']])
table.add_column(dictKeys[1], [data['Jumlah Attribute']])
table.add_column(dictKeys[2], ['\n'.join(data['Nama Attribute'])])
table.add_column(dictKeys[3], [data['Jumlah Attribute Dengan Tipe Angka']])
print(table)
```
